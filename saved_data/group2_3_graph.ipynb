```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainbow = np.load('training_group_2/training_rewards.npy')
dqn_pr = np.load('training_group_3/training_rewards.npy')
test_frequency = 5
print(f"Rainbow was trained for {test_frequency*len(rainbow)} epochs")
print(f"DQN PR was trained for {test_frequency*len(dqn_pr)} epochs")
```

```python
fig = plt.figure(figsize = (10,10))
plt.plot(test_frequency*np.arange(len(rainbow[:500])), rainbow[:500], label = f"Rainbow, final score = {round(np.mean(rainbow[-20:]),2)}")
plt.plot(test_frequency*np.arange(len(dqn_pr[:500])), dqn_pr[:500], label = f"Double DQN with PR, final score = {round(np.mean(dqn_pr[-20:]),2)}")
plt.legend()
plt.title('Training curve of Rainbow and Double DQN with PR on small Hanabi', fontsize = 15)
plt.ylabel('Average Score', fontsize = 15)
plt.xlabel('Epoch', fontsize =15)
plt.savefig('rainbow_vs_ddqnPR_small_hanabi.png')
plt.show()
```
