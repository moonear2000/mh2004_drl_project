```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
layers = [128,256]
testing_freq = 3
hand_sizes = [2,3,4]
data = {}
for l in layers:
    for s in hand_sizes:
        data[f'{l}_{s}'] = np.load(f'results/hanabi_small_{l}_size{s}/training_rewards.npy')
```

```python
# plot the training curves by layer
fig = plt.figure(figsize = (10,10))
l = 128
for s in hand_sizes:
    d = data[f'{l}_{s}']
    plt.plot(testing_freq*np.arange(len(d)),d, label = f'Hand size = {s}, Average score = {round(np.mean(d[-20:]),2)}')
plt.legend()
plt.title(f'Training curve of RDQN with {[l,l]} hidden layers on small Hanabi', fontsize = 15)
plt.xlabel('Epoch', fontsize = 15)
plt.ylabel('Average score', fontsize = 15)
plt.savefig(f'small_{l}_training_curves_RDQN.png')
plt.show()
```

```python
# plot the training curves by layer
fig = plt.figure(figsize = (10,10))
l = 256
for s in hand_sizes:
    d = data[f'{l}_{s}']
    plt.plot(testing_freq*np.arange(len(d)),d, label = f'Hand size = {s}, Average score = {round(np.mean(d[-20:]),2)}')
plt.legend()
plt.title(f'Training curve of RDQN with {[l,l]} hidden layers on small Hanabi', fontsize = 15)
plt.xlabel('Epoch', fontsize = 15)
plt.ylabel('Average score', fontsize = 15)
plt.savefig(f'small_{l}_training_curves_RDQN.png')
plt.show()
```
