```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hints = [4,5]
```

```python
data = {}
for h in hints:
    data[h] = np.load(f'results/hint{h}/training_rewards.npy')
data[3] = np.load(f'../hanabi_small/results/hanabi_small_256_size3/training_rewards.npy')
```

```python
hints = [3,4,5]
fig = plt.figure(figsize = (10,10))
for h in hints:
    d = data[h]
    plt.plot(np.arange(len(d))*3, d, label = f'hints = {h}, average score = {round(np.mean(d[-20:]),2)}')
plt.legend()
plt.title('Training curve of [256,256] RDQN training on small Hanabi with hand size 3 \n and different number of hints', fontsize = 15)
plt.xlabel('Epochs', fontsize = 15)
plt.ylabel('Average score (out of 10)', fontsize = 15)
plt.savefig('small_345hints_rdqn.png')
plt.show()
```
